This <a href="https://www.evaluationengineering.com/home/article/13000306/filters-eliminate-the-aliasing-phenomenon">article</a> introduces a simple way to remove aliasing. Implement the method in the article and show the results by triangle wave.

```python
import sys
sys.path.append('../ThinkDsp')
import thinkdsp
```

```python
def half_framerate(w):
    fr = w.framerate/2
    l = round(len(w.ys)/2)
    ys = [0.0]*l
    ts = [0.0]*l
    for i in range(l):
        ys[i] = w.ys[i*2]
        ts[i] = w.ts[i*2]
    w2 = thinkdsp.Wave(ys,ts,fr)
    return(w2)
```

```python
wave = thinkdsp.read_wave("triangle_mod.wav")
sp = wave.make_spectrum()
```

Implement the method here:

```python
w = sp.make_wave()
w2 = half_framerate(w)
w2.make_audio()
```

```python
seg = w2.segment(0,0.01)
seg.plot()
```

```python
sp2 = w2.make_spectrum()
sp2.plot()
```
