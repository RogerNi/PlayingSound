It is possible to hide message in sound by manipulating the spectrogram. Some examples can be found in <a href="https://twistedsifter.com/2013/01/hidden-images-embedded-into-songs-spectrographs/">here</a>. The following is an example to draw a trapezium in spectrogram.

```python
%matplotlib inline

import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot
```

```python
signal1 = thinkdsp.Chirp(start=200, end=1400)
wave1 = signal1.make_wave(duration=2)
```

```python
signal2 = thinkdsp.Chirp(start=1400, end=200)
wave2 = signal2.make_wave(duration=2)
```

```python
wave3 = thinkdsp.sin_wave(freq=200, duration=6)
```

```python
wave4 = thinkdsp.sin_wave(freq=1400, duration=2)
```

```python
wave5 = thinkdsp.rest(2)
```

```python
wave = wave1 | wave4 | wave2
wave = wave + wave3
wave = wave5 | wave | wave5
wave.make_audio()
```

```python
wave.plot()
```

```python
sg = wave.make_spectrogram(seg_length=1024)
sg.plot(high=2500)
```

Put the letter W in the frequency range 1700Hz - 2100Hz. Put the letter A in the frequency range 1200Hz - 1600Hz. Put the letter V in the frequency range 700Hz - 1100Hz. Put the letter E in the frequency range 200Hz - 600Hz. 

### Step 1: Make the wave for the letter W in spectrogram in the frequency range 1700Hz - 2100Hz


```python
signal1 = thinkdsp.Chirp(start=2100, end=1700)
wave1 = signal1.make_wave(duration=2)

signal2 = thinkdsp.Chirp(start=1700, end=1900)
wave2 = signal2.make_wave(duration=2)

signal3 = thinkdsp.Chirp(start=1900, end=1700)
wave3 = signal3.make_wave(duration=2)

signal4 = thinkdsp.Chirp(start=1700, end=2100)
wave4 = signal4.make_wave(duration=2)

waveW = wave1 | wave2 | wave3 | wave4
waveW.make_audio()
```

```python
sgW = waveW.make_spectrogram(seg_length=1024)
sgW.plot(high=2500)
```

### Step 2: Make the wave for the letter A in spectrogram in the frequency range 1200Hz - 1600Hz

```python
signal5 = thinkdsp.Chirp(start=1200, end=1600)
wave5 = signal5.make_wave(duration=4)

signal6 = thinkdsp.Chirp(start=1600, end=1200)
wave6 = signal6.make_wave(duration=4)

wave7 = thinkdsp.sin_wave(freq=1400, duration=4)

wave8 = thinkdsp.rest(2)

wave9 = wave5|wave6
wave10 = wave8|wave7|wave8

waveA = wave9 + wave10
waveA.make_audio()
```

```python
sgA = waveA.make_spectrogram(seg_length=1024)
sgA.plot(high=2500)
```

### Step 3: Make the wave for the letter V in spectrogram in the frequency range 700Hz - 1100Hz

```python
signal11 = thinkdsp.Chirp(start=1100, end=700)
wave11 = signal11.make_wave(duration=4)

signal12 = thinkdsp.Chirp(start=700, end=1100)
wave12 = signal12.make_wave(duration=4)

waveV = wave11|wave12
waveV.make_audio()
```

```python
sgV = waveV.make_spectrogram(seg_length=1024)
sgV.plot(high=2500)
```

### Step 4: Make the wave for the letter E in spectrogram in the frequency range 200Hz - 600Hz

```python
signal13 = thinkdsp.Chirp(start=500, end=600)
wave13 = signal13.make_wave(duration=8)

signal14 = thinkdsp.Chirp(start=500, end=400)
wave14 = signal14.make_wave(duration=5)

signal15 = thinkdsp.Chirp(start=300, end=400)
wave15 = signal15.make_wave(duration=5)

signal16 = thinkdsp.Chirp(start=300, end=200)
wave16 = signal16.make_wave(duration=8)

waveE = wave13 + wave14 + wave15+ wave16
waveE.make_audio()
```

```python
sgE = waveE.make_spectrogram(seg_length=1024)
sgE.plot(high=2500)
```

### Step 5: Mix waves of W, A, V, and E

```python
wave = waveW + waveA + waveV + waveE
wave.make_audio()
```

```python
sg = wave.make_spectrogram(seg_length=1024)
sg.plot(high=2300)
```
