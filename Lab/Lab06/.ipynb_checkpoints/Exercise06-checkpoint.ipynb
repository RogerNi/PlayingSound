The Python script below shows the way to plot a line in 3D.

```python
%matplotlib notebook
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection="3d")

z_line = np.arange(0, 15, 0.001)
x_line = np.log(z_line + 1)
y_line = np.power(z_line - 5,3)

ax.plot3D(x_line, y_line, z_line)

plt.show()
```

Implement 3D harmonographs for the frequency ratios in the major chord (4:5:6) and the minor chord (10:12:15). You should see 2D harmonograph in different projections.

```python
amp = 100
def three_d_harmono(fx, fy, fz):
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    t = np.arange(0, 15, 0.001)
    x = amp * np.sin(2*np.pi * fx * t)
    y = amp * np.sin(2*np.pi * fy * t)
    z = amp * np.sin(2*np.pi * fz * t)
    
    ax.plot3D(x, y, z)

    plt.show()
```

```python
three_d_harmono(4, 5 ,6)
```

```python
three_d_harmono(10, 12, 15)
```
