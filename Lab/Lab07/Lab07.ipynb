## Functions to generate scales

```python
def equalTempScale(stdA):
    fe = [0]*89
    fe[0] = 0
    for i in range(88):
        fe[i+1] = stdA * 2**((i-48)/12)
    return(fe)

etScale = equalTempScale(A4)
```

```python
def pythScale(start_index, stdA):
    x = [0]*12
    x[0] = 1
    for i in range(11):
        r = x[i] * 3/2
        if r > 2 :
            r = r/2
        x[i+1] = r    
    x.sort()
    
    fp = [0]*89
    fp[0] = 0
    freq = soundSyn.etScale[start_index]
    
    for i in range(12):
        fp[start_index+i] = freq * x[i]
        
    for p in range(5):
        for q in range(12):
            fp[start_index + q + (p+1)*12] = fp[start_index + q] * 2**(p+1)
    return(fp)

pyScale =  pythScale(12, A4) # G1
```

## Equal temperament tuning

```python
ssET = soundSyn.SoundSyn()
ssET.set_scale(etScale)
ssET.set_file("bwvAnhII114.txt")
wave = ssET.make_wave()
wave.make_audio()
```

## Equal temperament tuning with different partial amplitude ratios and bpm

```python
ssET.set_bpm(200)
ssET.set_partial([0.5, 0.5, 0.5, 0.3, 0.3, 0.2, 0.2, 0.1, 0.8])
wave = ssET.make_wave()
wave.make_audio()
```

## Pythagorean tuning

```python
ssPY = soundSyn.SoundSyn()
ssPY.set_scale(pyScale)
ssPY.set_file("bwvAnhII114.txt")
wave = ssPY.make_wave()
wave.make_audio()
```
