# Exercise 07

Implement a simple melody generator. 

The variable $\texttt{noteList}$ is a list of number of half tones to base note of the key of notes. The variable $\texttt{lengList}$ is a list of (relative) durations of notes. You can find frequencies of notes in $\texttt{etScale}$. 

The melody is in key G. Thus, the frequency of base note can be found in $\texttt{etScale[47]}$.

```python
import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot
```

```python
def equalTempScale(stdA):
    fe = [0]*89
    fe[0] = 0
    for i in range(88):
        fe[i+1] = stdA * 2**((i-48)/12)
    return(fe)

etScale = equalTempScale(440)
```

```python
noteList = [7,0,2,4,5,7,0,0,9,5,7,9,11,12,0,0]
lengList = [2,1,1,1,1,2,2,2,2,1,1,1,1,2,2,2]
```

Implement the generator here:

```python
l = len(noteList)
wave = thinkdsp.rest(0.2)
base_G = 47
for note, length in zip(noteList, lengList):
    f = etScale[note + base_G]
    w = thinkdsp.sin_wave(f, length)
    w.apodize()
    wave = wave | w
```

Test the result:

```python
wave.make_audio()
```
